Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how
they may impact your analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#importing data set
df = pd.read_excel("flight_price.xlsx")

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(y='Price', data=df)
plt.title('Distribution of Flight Prices')
plt.ylabel('Price')
plt.show()

# Calculate some statistics
q1 = df['Price'].quantile(0.25)
q3 = df['Price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]
print(f"Number of outliers: {len(outliers)}")
print(f"Percentage of outliers: {(len(outliers) / len(df) * 100):.2f}%")
print("\
Top 5 outliers:")
print(outliers.sort_values('Price',ascending=False).head())
```

```python
# Calculate statistics with and without outliers
q1 = df['Price'].quantile(0.25)
q3 = df['Price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

df_without_outliers = df[(df['Price'] >= lower_bound) & (df['Price'] <= upper_bound)]

print("Statistics with outliers:")
print(df['Price'].describe())
print("\
Statistics without outliers:")
print(df_without_outliers['Price'].describe())

print("\
Top 5 outliers:")
outliers = df[df['Price'] > upper_bound].sort_values('Price', ascending=False).head()
print(outliers[['Airline', 'Price']])

# Calculate percentage difference in mean and standard deviation
mean_diff = (df['Price'].mean() - df_without_outliers['Price'].mean()) / df_without_outliers['Price'].mean() * 100
std_diff = (df['Price'].std() - df_without_outliers['Price'].std()) / df_without_outliers['Price'].std() * 100

print(f"\
Percentage increase in mean due to outliers: {mean_diff:.2f}%")
print(f"Percentage increase in standard deviation due to outliers: {std_diff:.2f}%")
```

1)The maximum price drops from 79,512 to 23,001 when removing outliers, dramatically reducing the range of prices.
2)Percentage increase in mean due to outliers: 2.13%
3)Percentage increase in standard deviation due to outliers: 13.33%

Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
the relationship.

```python
# Remove 'M' and 'k' from the 'Size' column and convert to numeric
# For this analysis, we will ignore the 'M' and 'k' suffixes and treat all sizes as numeric values
size_numeric = df1['Size'].replace({'M': '', 'k': ''}, regex=True)

# Convert to numeric, treating errors as NaN
size_numeric = pd.to_numeric(size_numeric, errors='coerce')

# Update the dataframe with the cleaned size values

df1['Size'] = size_numeric

# Drop rows with missing values in 'Size' or 'Rating'
df1_cleaned = df1.dropna(subset=['Size', 'Rating'])

# Perform analysis on the cleaned data
# For example, we can calculate the average rating and size
average_rating = df1_cleaned['Rating'].mean()
average_size = df1_cleaned['Size'].mean()

# Count the number of apps in each category
category_counts = df1_cleaned['Category'].value_counts()

# Display the results
print(f"Average Rating: {average_rating}")
print(f"Average Size: {average_size}")
print("Category Counts:")
print(category_counts.head())

print("Analysis completed.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plot
plt.style.use('seaborn')

# Create the scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df1_cleaned, x='Size', y='Rating', alpha=0.5)

# Set labels and title
plt.xlabel('App Size')
plt.ylabel('Rating')
plt.title('App Size vs Rating in Google Play Store')

# Add a trend line
sns.regplot(data=df1_cleaned, x='Size', y='Rating', scatter=False, color='red')

# Show the plot
plt.tight_layout()
plt.show()

print("Scatter plot created.")

# Calculate correlation coefficient
correlation = df1_cleaned['Size'].corr(df1_cleaned['Rating'])
print(f"Correlation coefficient between Size and Rating: {correlation}")
```
